```java
public class Arme{
    private String nom;
    private int degats;
    private double attackSpeed;

    public Arme(String nom, int degats, double attackSpeed){
        this.nom = nom;
        this.degats = degats;
        this.attackSpeed = attackSpeed;
    }

    public int getDegats(){
        return degats;
    }

    public double getAttackSpeed(){
        return (double)attackSpeed;
    }

    public String getNom(){
        return nom;
    }
}
```

```java
public class Armure{
    private String nom;
    private int resistance;
    private int weight;

    public Armure(String nom, int resistance, int weight){
        this.nom = nom;
        this.resistance = resistance;
        this.weight = weight;
    }

    public int getResistance(){
        return resistance;
    }

    public String getNom(){
        return nom;
    }

    public int getWeight(){
        return weight;
    }
}
```

```java
public class Potion{
    private String nom;

    public Potion(String nom){
        this.nom = nom;
    }

    public String getNom(){
        return nom;
    }
}
```

```java
import java.util.List;
import java.util.ArrayList;

public class Inventaire{
    private List<Potion> potions;

    public Inventaire(){
        this.potions = new ArrayList<Potion>();
    }

    public void addPotion(Potion potion){
        this.potions.add(potion);
    }

    public List<Potion> getPotions(){
        return potions;
    }

    public void removePotion(Potion potion){
        this.potions.remove(potion);
    }
}
```

```java
import java.util.Random;

public class Champion{
    private String nom;
    private int vie;
    private Arme arme;
    private Armure armure;
    private int xp;
    private int level;
    private Inventaire inventaire;
    private Arme oldWeapon;

    public Champion(String nom, int vie, Arme arme, Armure armure){
        this.nom = nom;
        this.vie = 100;
        this.arme = arme;
        this.oldWeapon = arme;
        this.armure = armure;
        this.xp = 0;
        this.level = 0;
        this.inventaire = new Inventaire();
    }

    public Arme getOldWeapon(){
        return oldWeapon;
    }

    public String getNom(){
        return nom;
    }

    public void setNom(String nom){
        this.nom = nom;
    }

    public int getVie(){
        return vie;
    }

    public void setVie(int vie){
        this.vie = vie;
    }

    public Arme getArme(){
        return arme;
    }

    public void setArme(Arme arme){
        this.arme = arme;
    }

    public Armure getArmure(){
        return armure;
    }

    public void setArmure(Armure armure){
        this.armure = armure;
    }

    public int getLevel(){
        return level;
    }

    public int getXp(){
        return xp;
    }

    public void gagnerXp(int xpGagne){
        this.xp += xpGagne;
        if(this.xp >= 100){
            this.level += 1;
            this.xp = this.xp - 100;
        }
    }

    public void takeDamage(Arme arme){
        if (arme.getNom().equals("Gambling Blade")){
            Random rand = new Random();
            int chance = rand.nextInt(150)-25;
            this.vie -= (chance - this.armure.getResistance());
        }else{
            this.vie -= (arme.getDegats() - this.armure.getResistance());
        }
    }

    public float getMoveSpeed(){
        return (float)this.arme.getAttackSpeed() / (this.armure.getWeight()/10);
    }

    public Inventaire getInventaire(){
        return inventaire;
    }

    public void UsePotion(Potion potion){
        if(this.inventaire.getPotions().contains(potion)){
            if(potion.getNom() == "Health Potion"){
                this.vie += 20;
            }
            else if(potion.getNom() == "Gambling Potion"){
                Random rand = new Random();
                int chance = rand.nextInt(100);
                if(chance < 25){
                    this.vie = 0;
                } else if (chance > 25 && chance < 40){
                    this.arme = new Arme("Gambling Blade", 50, 0.1);
                }else{
                    this.vie += 15;
                }
            }
            this.inventaire.removePotion(potion);
        }
    }
}
```

```java
public class Combat{
    public void attaquer(Champion attaquant, Champion defenseur){
        Arme armeAttaquant = attaquant.getArme();
        Arme armeDefenseur = defenseur.getArme();
        if (armeAttaquant.getAttackSpeed() < armeDefenseur.getAttackSpeed()){
            attaquant.takeDamage(armeDefenseur);
        }
        else{
            defenseur.takeDamage(armeAttaquant);
        }
    }

    public boolean aliveVerification(Champion champion){
        if(champion.getVie() <= 0){
            return false;
        }
        return true;
    }
}
```
