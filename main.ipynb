Nous avons commencé par créer une classe champion qui représentera le joueur. Nous avons de nombreux paramètres privés décrivant ce champion tel que son nom, son nombre de points de vie, son arme, son armure, son expérience, son niveau actuel, son inventaire et enfin, son arme au début du combat. 
le constructeur nous permet de créer notre héro de façon personnalisée. les paramètres que l'on ne peut pas choisir sont : 
- Ses points de vie (fixé à 100)
- Son expérience (Commence à 0)
- Son niveau (Commence à 0 également)

Pour le reste, l'utilisateur est libre de choisir. 
Nous avons ensuite des fonctions get et set afin de récupérer les informations du personnages en jeu et/ou de les changer (par exemple lors d'un combat, nous allons utiliser getVie afin de récuperer le nombre de points de vie, puis utiliser setVie afin de modifer cette valeur après des dégats reçus). 

Nous avons également faot plusieurs fonctionnalités tel que gagnerXP() qui permet au joueur de gagner de l'expérience au combat, et de gagner un niveau si l'expérience est supérieure à 100. elle réinitialise ensuite l'xp.

Nous avons ensuite "takeDamage" qui permet de calculer les dégâts reçus au combat. cette fonction prend simplement la vie actuelle du personnage, retire le nombre de dégat de l'arme ennemie een comptant l'armure du personnage. Il y a également un cas spécial si l'arme utilisée est la "gambling blade" où les dégats fait son aléatoires. 

Finalement, nous avons une méthode "usePotion" montrant une autre fonctionnalité du jeu, où le joueur peut choisir une potion à boire plutôt que d'attaquer. Cette méthode permet de voir si la potion utilisée est une potion de vie, dans ce cas elle rend 20 points de vie, ou si c'es une potion de GAMBLING, où le joueur a donc 25% de chance de mourir, 15% de chance d'avoir une arme spéciale, et 60% de chance de régénérer 15 PV.



```java
import java.util.Random;

public class Champion{
    private String nom;
    private int vie;
    private Arme arme;
    private Armure armure;
    private int xp;
    private int level;
    private Inventaire inventaire;
    private Arme oldWeapon;

    public Champion(String nom, int vie, Arme arme, Armure armure){
        this.nom = nom;
        this.vie = 100;
        this.arme = arme;
        this.oldWeapon = arme;
        this.armure = armure;
        this.xp = 0;
        this.level = 0;
        this.inventaire = new Inventaire();
    }

    public Arme getOldWeapon(){
        return oldWeapon;
    }

    public String getNom(){
        return nom;
    }

    public void setNom(String nom){
        this.nom = nom;
    }

    public int getVie(){
        return vie;
    }

    public void setVie(int vie){
        this.vie = vie;
    }

    public Arme getArme(){
        return arme;
    }

    public void setArme(Arme arme){
        this.arme = arme;
    }

    public Armure getArmure(){
        return armure;
    }

    public void setArmure(Armure armure){
        this.armure = armure;
    }

    public int getLevel(){
        return level;
    }

    public int getXp(){
        return xp;
    }

    public void gagnerXp(int xpGagne){
        this.xp += xpGagne;
        if(this.xp >= 100){
            this.level += 1;
            this.xp = this.xp - 100;
        }
    }

    public void takeDamage(Arme arme){
        if (arme.getNom().equals("Gambling Blade")){
            Random rand = new Random();
            int chance = rand.nextInt(150)-25;
            this.vie -= (chance - this.armure.getResistance());
        }else{
            this.vie -= (arme.getDegats() - this.armure.getResistance());
        }
    }

    public float getMoveSpeed(){
        return (float)this.arme.getAttackSpeed() / (this.armure.getWeight()/10);
    }

    public Inventaire getInventaire(){
        return inventaire;
    }

    public void UsePotion(Potion potion){
        if(this.inventaire.getPotions().contains(potion)){
            if(potion.getNom() == "Health Potion"){
                this.vie += 20;
            }
            else if(potion.getNom() == "Gambling Potion"){
                Random rand = new Random();
                int chance = rand.nextInt(100);
                if(chance < 25){
                    this.vie = 0;
                } else if (chance > 25 && chance < 40){
                    this.arme = new Arme("Gambling Blade", 50, 0.1);
                }else{
                    this.vie += 15;
                }
            }
            this.inventaire.removePotion(potion);
        }
    }
}
```

Cette classe est la classe mère des différentes armes du jeu. Elle permet de créer une arme facilement basé sur son nom, ses dégâts, et sa vitesse d'attaque. Nous avons ensuite simplement trois fonctions get pour récupérer ces statistiques pour les combats. Plusieurs armes découlent de cette classe telle que "Fist", une arme utilisant donc la classe Arme afin de créer une arme de dégât 2 et vitesse d'attaque 10. 

```java
public class Arme{
    private String nom;
    private int degats;
    private double attackSpeed;

    public Arme(String nom, int degats, double attackSpeed){
        this.nom = nom;
        this.degats = degats;
        this.attackSpeed = attackSpeed;
    }

    public int getDegats(){
        return degats;
    }

    public double getAttackSpeed(){
        return (double)attackSpeed;
    }

    public String getNom(){
        return nom;
    }
}
```

```java
public class Fist extends Arme{
    public Fist(){
        super("Fist", 2, 10);
    }
}
```

```java
public class Epee extends Arme{
    public Epee(){
        super("Epee", 15, 1.5);
    }
}
```

```java
public class Hache extends Arme{
    public Hache(){
        super("Hache", 20, 1);
    }
}
```

```java
public class Log extends Arme{
    public Log(){
        super("Log", 35, 0.5);
    }
}
```

```java
public class Cancer extends Arme{
    public Cancer(){
        super("Cancer", 10000, 0.2);
    }
}
```

Nous avons également créer une classe mère Armure qui permettra au joueur de résister une partie des dégats, en sacrifiant sa vitesse. de la même manière que la classe Arme, elle se repose sur 3 paramètres (ici le nom, la résistance, et le poids), avec 3 méthodes get pour avoir ses statistiques au combat. 

Plusieurs classe en découle tel que "WoodenArmour" avec une résistance de 5 et un poids de 50.

```java
public class Armure{
    private String nom;
    private int resistance;
    private int weight;

    public Armure(String nom, int resistance, int weight){
        this.nom = nom;
        this.resistance = resistance;
        this.weight = weight;
    }

    public int getResistance(){
        return resistance;
    }

    public String getNom(){
        return nom;
    }

    public int getWeight(){
        return weight;
    }
}
```

```java
public class Naked extends Armure{
    public Naked(){
        super("Naked", 0, 0);
    }
}
```

```java
public class WoodenArmour extends Armure{
    public WoodenArmour(){
        super("Wooden Armour", 5, 50);
    }
}
```

```java
public class IronArmour extends Armure{
    public IronArmour(){
        super("Iron Armour", 10, 100);
    }
}
```

Les potions sont gérées par la classe Potion prenant uniquement leur nom en paramètres. Il n'y a qu'une méthode get car le nom est la seule chose utilisée pour déterminer quelle potion est utilisée. 

```java
public class Potion{
    private String nom;

    public Potion(String nom){
        this.nom = nom;
    }

    public String getNom(){
        return nom;
    }
}
```

Cette classe utilise une liste de potion afin de laisser le joueur choisir quelle chose utiliser au combat. Cet inventaire n'a pas de limite de taille et ne contient que des potions. Nous avons une méthode addPotion afin de rajouter des potions à son inventaire, et une fonction removePotion afin de supprimer une potion après utilisation.

```java
import java.util.List;
import java.util.ArrayList;

public class Inventaire{
    private List<Potion> potions;

    public Inventaire(){
        this.potions = new ArrayList<Potion>();
    }

    public void addPotion(Potion potion){
        this.potions.add(potion);
    }

    public List<Potion> getPotions(){
        return potions;
    }

    public void removePotion(Potion potion){
        this.potions.remove(potion);
    }
}
```

La classe combat permet au joueur et à son ennemi de combattre afin de gagner de lxp. La méthode attaquer prends en paramètre les deux personnages attaquant et défenseur. Il récupère leurs armes et regarde quel personnage a la vitesse d'attaque la plus haute. Cette personne attaque donc en premier. En cas d'égalité, l'attaquant attaque toujours en premier. Après cette vérification, nous utilisons la fonction takeDamage depuis Champion pour réduire les points de vie des personnages. 

Finalement, nous avons une méthode pour vérifier si les personnages sont toujours en vie, c'est à dire si leur vie a atteint la valuer 0 ou en dessous. Cela renvoie un simple boolean, et est utilisé pour mettre fin au combat. 

```java
public class Combat{
    public void attaquer(Champion attaquant, Champion defenseur){
        Arme armeAttaquant = attaquant.getArme();
        Arme armeDefenseur = defenseur.getArme();
        if (armeAttaquant.getAttackSpeed() < armeDefenseur.getAttackSpeed()){
            attaquant.takeDamage(armeDefenseur);
        }
        else{
            defenseur.takeDamage(armeAttaquant);
        }
    }

    public boolean aliveVerification(Champion champion){
        if(champion.getVie() <= 0){
            return false;
        }
        return true;
    }
}
```

Le programme commence par créer plusieurs ennemis. Ensuite, il demande au joueur de choisir un nom, une arme et une armure. Une fois le joueur créé, il enchaîne des combats contre des ennemis choisis selon son niveau. À chaque tour, le joueur peut attaquer ou afficher les statistiques. S’il gagne un combat, il gagne de l’XP et continue. S’il meurt, la partie s’arrête. S’il survit à tous les combats, le jeu affiche ses points de vie restants.

```java
import java.util.Scanner;

Champion ennemi_1 = new Champion("Goblin", 50, new Epee(), new Naked());
Champion ennemi_2 = new Champion("Orc", 80, new Hache(), new WoodenArmour());
Champion ennemi_3 = new Champion("Troll", 120, new Log(), new Naked());
Champion ennemi_4 = new Champion("Kawaleck", 200, new Cancer(), new IronArmour());

Scanner myObj = new Scanner(System.in);

System.out.println("Enter username : ");
String userName = myObj.nextLine();

while (userName.isEmpty()) {
    System.out.println("Username cannot be empty. Enter username : ");
    userName = myObj.nextLine();
}
System.out.println("Enter weapon choice (1. Fist \n 2. Epee \n 3. Hache : \n)");

String weaponChoice = myObj.nextLine();
int weaponChoiceInt = Integer.parseInt(weaponChoice);

while ((weaponChoiceInt < 1 || weaponChoiceInt > 3)) {
    System.out.println("Invalid weapon choice. Enter weapon choice (1. Fist, 2. Epee, 3. Hache :)");
    weaponChoice = myObj.nextLine();
    weaponChoiceInt = Integer.parseInt(weaponChoice);
}

Arme playerWeapon = new Fist();
switch (weaponChoiceInt) {
    case 1:
        playerWeapon = new Fist();
        break;
    case 2:
        playerWeapon = new Epee();
        break;
    case 3:
        playerWeapon = new Hache();
}

System.out.println("Choisissez votre armure : \n 1. Naked \n 2. WoodenArmour \n 3. IronArmour : \n");

String armorChoice = myObj.nextLine();
int choiceInt = Integer.parseInt(armorChoice);

while (choiceInt != 1 && choiceInt != 2 && choiceInt != 3) {
    System.out.println(
            "Invalid armor choice. Choisissez votre armure : \n 1. Naked \n 2. WoodenArmour \n 3. IronArmour : \n");
    armorChoice = myObj.nextLine();
    choiceInt = Integer.parseInt(armorChoice);
}

Armure playerArmor = new Naked();
switch (choiceInt) {
    case 1:
        playerArmor = new Naked();
        break;
    case 2:
        playerArmor = new WoodenArmour();
        break;
    case 3:
        playerArmor = new IronArmour();
}

Champion player = new Champion(userName, 100, playerWeapon, playerArmor);

Combat combat = new Combat();
Champion[] ennemis = new Champion[]{ennemi_1, ennemi_2, ennemi_3, ennemi_4};

java.util.Random rand = new java.util.Random();

while (combat.aliveVerification(player)) {
    Champion ennemi;
    int lvl = player.getLevel();
    double r = rand.nextDouble();

    if (lvl < 5) {
        ennemi = ennemis[0];
    } else if (lvl >= 5 && lvl < 10) {
        ennemi = rand.nextBoolean() ? ennemis[0] : ennemis[1];
    } else if (lvl >= 10 && lvl < 20) {
        if (r < 0.75) {
            ennemi = rand.nextBoolean() ? ennemis[0] : ennemis[1];
        } else {
            ennemi = ennemis[2];
        }
    } else {
        if (r < 0.70) {
            ennemi = rand.nextBoolean() ? ennemis[0] : ennemis[1];
        } else if (r < 0.95) {
            ennemi = ennemis[2];
        } else {
            ennemi = ennemis[3];
        }
    }

    System.out.println("\n--- Nouveau combat : " + ennemi.getNom() + " ---");
    while (combat.aliveVerification(player) && combat.aliveVerification(ennemi)) {
        System.out.println("\nChoisissez une action :");
        System.out.println("1. Attaquer");
        System.out.println("2. Afficher statistiques");

        String action = myObj.nextLine().trim();
        if (action.equals("1")) {
            combat.attaquer(player, ennemi);
        } else if (action.equals("2")) {
            System.out.println(player.getNom() + " - Vie: " + player.getVie() + " | Arme: " + player.getArme().getNom() + " | Armure: " + player.getArmure().getNom() + " | XP: " + player.getXp() + " | Level: " + player.getLevel());
            System.out.println(ennemi.getNom() + " - Vie: " + ennemi.getVie() + " | Arme: " + ennemi.getArme().getNom() + " | Armure: " + ennemi.getArmure().getNom());
            continue;
        } else {
            System.out.println("Action invalide, réessayez.");
            continue;
        }

        System.out.println("Après l'échange : " + player.getNom() + " Vie = " + player.getVie() + " | " + ennemi.getNom() + " Vie = " + ennemi.getVie());
    }

    if (!combat.aliveVerification(player)) {
        System.out.println("\nVous êtes mort. Fin du jeu.");
        break;
    }

    if (!combat.aliveVerification(ennemi)) {
        System.out.println("\nVous avez vaincu " + ennemi.getNom() + " !");
        player.gagnerXp(50);
        System.out.println("Vous gagnez 50 XP. XP: " + player.getXp() + " | Level: " + player.getLevel());
        player.setArme(player.getOldWeapon());
    } else {
        System.out.println("Vous passez au prochain adversaire.");
    }
}

if (combat.aliveVerification(player)) {
    System.out.println("\nFin des affrontements. " + player.getNom() + " survit avec " + player.getVie() + " points de vie.");
}
```
