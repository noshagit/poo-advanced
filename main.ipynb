```java
public class Champion{
    private String nom;
    private int vie;
    private Arme arme;
    private Armure armure;
    private int xp;
    private int level;
    private Inventaire inventaire;
    private Arme oldWeapon;

    public Champion(String nom, int vie, Arme arme, Armure armure){
        this.nom = nom;
        this.vie = 100;
        this.arme = arme;
        this.oldWeapon = arme;
        this.armure = armure;
        this.xp = 0;
        this.level = 0;
        this.inventaire = new Inventaire();
    }

    public String getNom(){
        return nom;
    }

    public void setNom(String nom){
        this.nom = nom;
    }

    public int getVie(){
        return vie;
    }

    public void setVie(int vie){
        this.vie = vie;
    }

    public Arme getArme(){
        return arme;
    }

    public void setArme(Arme arme){
        this.arme = arme;
    }

    public Armure getArmure(){
        return armure;
    }

    public void setArmure(Armure armure){
        this.armure = armure;
    }

    public int getLevel(){
        return level;
    }

    public int getXp(){
        return xp;
    }

    public void gagnerXp(int xpGagne){
        this.xp += xpGagne;
        if(this.xp >= 100){
            this.level += 1;
            this.xp = this.xp - 100;
        }
    }

    public void gagnerLevel(){
        this.level += 1;
    }

    public int takeDamage(Champion champion, Arme arme){
        if (arme.getName().equals("Gambling Blade")){
            Random rand = new Random();
            int chance = rand.nextInt(150)-25;
            champion.vie -= (chance - this.armure.getResistance());
        }else{
            champion.vie -= (arme.getDegats() - this.armure.getResistance());
        }
    }

    public float getMoveSpeed(){
        return this.arme.getAttackSpeed() / (this.armure.getWeight()/10);
    }

    public Inventaire getInventaire(){
        return inventaire;
    }

    public void UsePotion(Potion potion){
        if(this.inventaire.getPotions().contains(potion)){
            if(potion.getType() == TypePotion.HEAL){
                this.vie += 20;
            }
            else if(potion.getType() == TypePotion.GAMBLING){
                Random rand = new Random();
                int chance = rand.nextInt(100);
                if(chance < 25){
                    this.vie = 0;
                } else if (chance > 25 && chance < 40){
                    this.arme = new Arme("Gambling Blade", 50, 0.1);
                }else{
                    this.vie += 15;
                }
            }
            this.inventaire.removePotion(potion);
        }
    }
}

public class Arme{
    private String nom;
    private int degats;
    private float attackSpeed;

    public Arme(String nom, int degats, float attackSpeed){
        this.nom = nom;
        this.degats = degats;
        this.attackSpeed = attackSpeed;
    }

    public int getDegats(){
        return degats;
    }

    public float getAttackSpeed(){
        return attackSpeed;
    }

    public String getNom(){
        return nom;
    }
}

public class Fist extends Arme{
    public Fist(){
        super("Fist", 2, 10);
    }
}

public class Epee extends Arme{
    public Epee(){
        super("Epee", 15, 1.5);
    }
}

public class Hache extends Arme{
    public Hache(){
        super("Hache", 20, 1);
    }
}

public class Log extends Arme{
    public Log(){
        super("Log", 35, 0.5);
    }
}

public class Cancer extends Arme{
    public Cancer(){
        super("Cancer", 10000, 0.2);
    }
}

public class Armure{
    private String nom;
    private int resistance;
    private int weight;

    public Arme(String nom, int resistance, int weight){
        this.nom = nom;
        this.resistance = resistance;
        this.weight = weight;
    }

    public int getResistance(){
        return resistance;
    }

    public String getNom(){
        return nom;
    }
}

public class Naked extends Armure{
    public Naked(){
        super("Naked", 0, 0);
    }
}

public class WoodenArmour extends Armure{
    public WoodenArmour(){
        super("Wooden Armour", 5, 50);
    }
}

public class IronArmour extends Armure{
    public IronArmour(){
        super("Iron Armour", 10, 100);
    }
}

public class Inventaire{
    private List<Potion> potions;

    public Inventaire(){
        this.potions = new ArrayList<Potion>();
    }

    public void addPotion(Potion potion){
        this.potions.add(potion);
    }

    public List<Potion> getPotions(){
        return potions;
    }

    public void removePotion(Potion potion){
        this.potions.remove(potion);
    }
}

public class Potion{
    private String nom;
    private enum TypePotion {HEAL, GAMBLING};

    public Potion(String nom, TypePotion type){
        this.nom = nom;
    }

    public String getNom(){
        return nom;
    }
}

public static class Combat{
    public void attaquer(Champion attaquant, Champion defenseur){
        Arme armeAttaquant = attaquant.getArme();
        Arme armeDefenseur = defenseur.getArme();
        if armeAttaquant.getAttackSpeed() < armeDefenseur.getAttackSpeed(){
            attaquant.takeDamage(armeDefenseur);
        }
        else{
            defenseur.takeDamage(armeAttaquant);
        }
    }

    public boolean aliveVerification(Champion champion){
        if(champion.getVie() <= 0){
            return false;
        }
        return true;
    }
}

public class Main{
    public static void main(String[] args){
        Champion ennemi_1 = new Champion("Goblin", 50, new Epee(), new Naked());
        Champion ennemi_2 = new Champion("Orc", 80, new Hache(), new WoodenArmour());
        Champion ennemi_3= new Champion("Troll", 120, new Log(), new Naked());
        Champion ennemi_4 = new Champion("Kawaleck", 200, new Cancer(), new IronArmour());


        Scanner myObj = new Scanner(System.in);

        System.out.println("Enter username : ");
        String userName = myObj.nextLine();
        while (userName.isEmpty()){
            System.out.println("Username cannot be empty. Enter username : ");
            userName = myObj.nextLine();
        }
        System.out.println("Enter weapon choice (1. Fist \n 2. Epee \n 3. Hache : \n)");
        int weaponChoice = myObj.nextInt();
        while (!(weaponChoice < 1 || weaponChoice > 3)){
            System.out.println("Invalid weapon choice. Enter weapon choice (1. Fist \n 2. Epee \n 3. Hache : \n)");
            weaponChoice = myObj.nextInt();
        }

        switch (weaponChoice) {
            case 1:
                Arme playerWeapon = new Fist();
                break;
            case 2:
                Arme playerWeapon = new Epee();
                break;
            case 3:
                Arme playerWeapon = new Hache();
        }

        System.out.println("Choisissez votre armure : \n 1. Naked \n 2. WoodenArmour \n 3. IronArmour : \n");
        int armorChoice = myObj.nextInt();

        while (armorChoice < 1 || armorChoice > 3){
            System.out.println("Invalid armor choice. Choisissez votre armure : \n 1. Naked \n 2. WoodenArmour \n 3. IronArmour : \n");
            armorChoice = myObj.nextInt();
        }

        switch (armorChoice) {
            case 1:
                Armure playerArmor = new Naked();
                break;
            case 2:
                Armure playerArmor = new WoodenArmour();
                break;
            case 3:
                Armure playerArmor = new IronArmour();
        }

        Champion player = new Champion(userName, 100, playerWeapon, playerArmor);
    }
}
```
